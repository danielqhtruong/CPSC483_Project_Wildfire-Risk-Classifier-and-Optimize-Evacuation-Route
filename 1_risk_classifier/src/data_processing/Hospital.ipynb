## LA County Hospital and Medical Centers Locations

```python
# read in Hospital locations
hospital_df = gpd.read_file("../../../data/raw/Hospitals_and_Medical_Centers.geojson")
hospital_df.head()
```

```python
hospital_df = hospital_df[["OBJECTID", "Name", "city", "geometry"]].copy()
hospital_df = hospital_df.rename(columns={"OBJECTID": "objectid", "Name": "name"})
hospital_df.head()
```

```python
hospital_df.to_file('../../../data/processed/hospitals.geojson')
```
