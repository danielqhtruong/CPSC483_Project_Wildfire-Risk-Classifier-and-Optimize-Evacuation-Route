## Fire Hydrants Locations accorss LA County

```python
# read in LACoFD station locations
hydrants = gpd.read_file("../../../data/raw/Fire_Hydrants_(DWP).geojson")
hydrants.head()
```

```python
hydrants = hydrants[["OBJECTID", "HYDRTS_ID", "geometry"]].copy()
hydrants = hydrants.rename(columns={"OBJECTID": "objectid", "HYDRTS_ID": "hydrantid"})
hydrants.head()
```

```python
# save dataframe
hydrants.to_file('../../../data/processed/Fire_Hydrants.geojson')
```
